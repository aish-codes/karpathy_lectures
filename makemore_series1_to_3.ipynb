```python
# initialize dict
# dict output is 'a' : 1, 'b' : 2, etc
b = {} # this is our character level dictionary that we'd use everywhere going forward
for w in words_list_filtered :
  chs = ["."] + list(w) + ["."]
  for ch1, ch2 in zip(chs, chs[1:]):
    bigram = (ch1, ch2)
    b[bigram] = b.get(bigram, 0) + 1

b_flat = sorted(set([ch for bigram in b.keys() for ch in bigram]))
```

```python
# creating a dict assigning each word an integer
wtoi = {w:i for i,w in enumerate(b_flat)} #
wtoi['.'] = 0

# reversed the dict
itow = {i:w for w,i in wtoi.items()}
```

```python
# initializing a tensor. if you're reading then try and answer why did we initialize it with a 27 x 27 matrix instead of a 26 x 26 matrix
import torch
torch.set_printoptions(precision=4,sci_mode=False)
N = torch.zeros((27,27), dtype=torch.int32)

# storing combination of each letter with the other for example aa, ab, ac & so on
for w in words_list_filtered :
  chs = ["."] + list(w) + ["."]
  for ix1, ix2 in zip(chs, chs[1:]):
    ix1 = wtoi[ix1]
    ix2 = wtoi[ix2]
    N[ix1, ix2] += 1
```

### So why did we create that N for?

1. In order to build a character level bigram model we need to calculate probabilities of each character occuring after a character
2. With N we have a matrix of 27 x 27 with combination of each letter with the other and their occurences

```python
# Generating words based on our matrix N
# This is a very crude bigram language model
# import library
import random

# pick a random starting index
n = random.randint(1, 26)
g = torch.Generator() # removed maual seed otherwise it was generating same words again and again

for _ in range(10):   # generate 10 random words
    word = []
    n = wtoi["."]   # always start from start token i.e. '.'
    while True:
        p = N[n].float()
        p /= p.sum()  # normalize
        ix = torch.multinomial(p, num_samples=1, replacement=True).item() # special torch fn for generating samples where p represents probabilities and it randomly picks a number
        letter = itow[ix]

        if letter == '.':   # end of word
            break

        word.append(letter)
        n = ix  # move to next state

    print(''.join(word)) # should produce absolute gutter atm
```

## What should be an ideal bigram model?

1. The best model would be certain in terms of its output of what comes next, i.e. the probability should be 1 fot he next letter.
2. Models are trained to minimize *loss functions* and we create a function to measure our model's capability. **We calculate the log of the probability of the combination of letters based on N**
3. The average negative of the this log likelihood is then used as our metric to measure our model performance


```python
# sample function for understanding
# optimizing the loss function over
P = N.float()
# read about the keepdims param and how it affects the sum function here
P /= P.sum(1,keepdims = True) # probability distribution for each letter combination
log_likelihood, n = 0, 0 # initialize

for w in words_list_filtered[:5] : # using sample of 5 words
  w = ["."] + list(w) + ["."]
  for ch1, ch2 in zip(w, w[1:]) :
    ix1 = wtoi[ch1]
    ix2 = wtoi[ch2]
    p = P[ix1, ix2]
    log_likelihood += torch.log(p)
    n += 1

nll = -log_likelihood # negative log likelihood
print(f'log likelihood = {log_likelihood}')
print(f'negative log likelihood = {nll}')
print(f'average negative log likelihood = {nll/n}')
```

## Implementing in pytorch

We created a sample code initially which we would reuse to recreate with torch below

```python
# implementing the above using torch now
xs, ys = [], []

for w in words_list_filtered:
  w = ["."] + list(w) + ["."]
  for ch1, ch2 in zip(w, w[1:]):
    ix1 = wtoi[ch1]
    ix2 = wtoi[ch2]
    xs.append(ix1)
    ys.append(ix2)

# here we have a mapping of x with y i.e. input and expected output where all letters are encoded by letters
xs, ys = torch.tensor(xs), torch.tensor(ys)
```

```python
import torch.nn.functional as F # for using ohe function
from torch import Generator

# forwards pass example
xenc = F.one_hot(xs, num_classes = 27).float() # using one hot encoding function to identify the letter
g = Generator().manual_seed(123)
W  = torch.randn((27, 27),generator = g, requires_grad=True) # initializing weights for layer 1

# view xenc to understand ohe
# xenc[1,:]
```

```python
# initialize learning rate
lr = 10 # play around with it to lower it even further

# importing library for calculating time taken to train the model
import time
start_time = time.time()

for _ in range(1000):
  # Doing the matmul
  logits = xenc @ W # [419720, 27] x [27, 27] =  [419720, 27]
  N = logits.exp() # exponentiating because mul with W might've resulted in neg values and we want probabilities in the next step
  # layer 1
  probs = N/N.sum(dim = 1, keepdims = True)
  ys = torch.tensor(ys, dtype=torch.long)   # ensure tensor type
  loss = -probs[torch.arange(xs.shape[0]), ys].log().mean()

  # backward pass
  W.grad =  None # setting it up to 0 for every loop
  loss.backward() # backpropogation

  # print(f'The nll is {loss}')

  W.data += -lr * W.grad

end_time = time.time()
final_time = end_time - start_time
print(f'The model took {final_time} seconds to run and the final loss nll is {loss}')
```

## Final results

```python
# generating words
# the above model can be optimized even more but we're sticking to it for the time being
for _ in range(10):   # generate 10 words
    word = []
    n = wtoi["."]   # start token

    while True:
        # one-hot encode current token
        xenc = F.one_hot(torch.tensor([n]), num_classes=27).float()
        logits = xenc @ W              # shape (1, 27)
        probs = logits.softmax(dim=1)  # turn into probabilities

        # sample next index
        ix = torch.multinomial(probs[0], num_samples=1).item()
        letter = itow[ix]

        if letter == ".":
            break

        word.append(letter)
        n = ix  # feed back as next input

    print("".join(word))
```

### What now?

The issues with our earlier code pice as I have highlighted earlier is that the cost to scale it would be very high

In this piece we would follow the below steps
1. increase context length from 2 to 5
2. we drop the N bigram probability matrix
3. encode our input
4. add a layer of weights *(W)* and bias *(b)*
5. pass ouput of 3rd pointer through a softmax func like tanh to generate probabilities for next character
6. instead of training the data on the whole dataset we'd use **mini batches** for updating the gradient which is extemely fast
7. instead of using logits and softmax we will use the cross_entropy function

```python
# declaring lists
X, Y = [], []

# karpathy in his lectures used a context length of 3
block_size = 5 # context length: how many characters do we take to predict the next one?

# building X, Y with context length 5
for w in words_list_filtered:
  context = [0] * block_size
  for ch in w + '.':
    ix = wtoi[ch]
    X.append(context)
    Y.append(ix)
    # print(''.join(itow[i] for i in context), '--->', itow[ix])
    context = context[1:] + [ix] # crop and append

# intializing X : input, Y: output tensors
X, Y = torch.tensor(X), torch.tensor(Y)
```

```python
# dividing our data
Xtr = X[:int(X.shape[0] * 0.8), :]
Ytr = Y[:int(Y.shape[0] * 0.8)]

Xte = X[int(X.shape[0] * 0.8):, :]
Yte = Y[int(Y.shape[0] * 0.8):]
```

```python
# initializing C
C = torch.randn(27,10)

# embedding
emb_tr = C[Xtr]
# C[X].shape # read about advanced indexing in torch and numpy to understand this
```

```python
# looping over the backward pass
g = torch.Generator().manual_seed(123)

# forward pass
W1 = torch.randn((50, 200), generator = g, requires_grad= True)
b1 = torch.randn(200,generator = g, requires_grad= True)
layer1 = torch.tanh(emb_tr.view(-1, 50) @ W1 + b1) # layer 1
W2 = torch.randn((200, 27), generator = g, requires_grad= True)
b2 = torch.randn(27, generator = g, requires_grad= True) # layer 2
layer2 = layer1 @ W2 + b2

# # create mini batch for input
# mini_batch = torch.randint(0, Xtr.shape[0], (32,))
# mini_batch_emb = C[Xtr[mini_batch]].view(-1,50) # 50 is selected inside view(-1,50) so that it can feed into W1

# initialize params
parameters = [W1, W2, b1, b2]

# list declaration
lri = []
lossi = []
```

```python
# backpropogation FTW
start_time = time.time()
for i in range(200000):

  # create mini batch for input
  mini_batch = torch.randint(0, Xtr.shape[0], (32,))
  mini_batch_emb = C[Xtr[mini_batch]].view(-1,50)

  # forward pass
  layer1 = torch.tanh(mini_batch_emb @ W1 + b1) # layer 1 # doesn't seem right to me as we're only pushing 32 examples here
  layer2 = layer1 @ W2 + b2

  # loss function calculation
  loss = F.cross_entropy(layer2, Ytr[mini_batch])

  # resetting gradient to 0
  for p in parameters:
    p.grad = None  # recommended over .zero_() for performance

  # backward pass
  loss.backward() # resulting in error

  # updating this to see our learning rate iteration and loss iteration
  lri.append(i)
  lossi.append(loss.item())

  # update the neural network
  for param in parameters :
    lr = 0.1 if i < 100000 else 0.01
    param.data += -lr * param.grad
end_time = time.time()
final_time = end_time - start_time

print(f'The model took {final_time} seconds to run and the final loss is {loss.item()}')
```

### Initialization and tanh saturation issue

We observe a hockey stick like loss function forming. This issue can be handled by initializing our *W* and *b* better.

Another problem that we see is with the histogram : we can clearly see there are a lot of absolute values which are greater than 0.99 or lesser than 0.99  

This is a problem because while calculating the gradients using backpropogation we will arrive at results with value 0.

To fix this issue we will fix the intialization and then look at the distribution of our batches


### The fix

In the below example we will scale down *W1* & *b1*  by multiplication with a smaller number. But how do we arrive at that number?  

The [batchnorm paper](https://arxiv.org/abs/1502.03167) suggests we multiply ***(gain/sqrt of fan_in)***  for arriving at this number for proper initialization rather than taking blind shots by multiplying arbitary numbers and then looking at the distributions

The gain depends upon the activation function in use

Pytorch has init functions which solves the initialization problem

```python
# forward pass
W1_test = W1.clone()
b1_test = b1.clone()

# create mini batch for input
mini_batch = torch.randint(0, Xtr.shape[0], (32,))
mini_batch_emb = C[Xtr[mini_batch]].view(-1,50)

# forward pass
layer1 = torch.tanh(mini_batch_emb @ W1_test.data.mul_(0.01) + b1_test.data.mul_(0.1))

# viz the graph
plt.figure(figsize = (10,20))
plt.imshow(layer1.abs() > 0.99, cmap = 'gray', interpolation = 'nearest')
```

```python
# check the distribution of the tanh function
plt.hist(layer1.view(-1).detach());
```

Implementing the initialization fix and batch normalisation for fixing the distribution and initialization issues at hand

```python
# the code fix
g = torch.Generator().manual_seed(123) # fixing the generations
emb_tr = C[Xtr]

# setting up bngain and bnbias
bngain = torch.ones((1, 200), requires_grad=True)
bnbias = torch.zeros((1, 200), requires_grad=True)

# this is to be used at the time of inference
mean_running = torch.ones((1, 200), requires_grad=True)
std_running = torch.zeros((1, 200), requires_grad=True)

# the weights at initalization needs to get multiplied with gain/fan_in ** 0.5
# gain varies depending upon the function used in the hidden layer
# for e.g. we have tanh = 5/3, relu = 1
W1 = torch.randn((50, 200), generator = g, requires_grad= True)
W1.data.mul_((5/3)/(50 ** 0.5)) # adjusting the gain for layer 1
b1 = torch.randn(200,generator = g, requires_grad= True)
W2 = torch.randn((200, 27), generator = g, requires_grad= True)
W2.data.mul_((5/3)/(200 ** 0.5)) # adjusting the gain for layer 2
b2 = torch.randn(27, generator = g, requires_grad= True)

# update params
parameters = [W1, W2, b1, b2, bngain, bnbias]

# list declaration
lri = []
lossi = []
lr = 0.1


start_time = time.time()
# looping over
for i in range(200000):
  # create mini batch for input
  # the requirement here is to normalise this batch and add the params which can be used for normalising this batch
  mini_batch = torch.randint(0, Xtr.shape[0], (32,))
  mini_batch_emb = C[Xtr[mini_batch]].view(-1,50)

  # forward pass
  # phase 1
  preact = mini_batch_emb @ W1 + b1
  meani = preact.mean(0,keepdim= True)
  stdi = preact.std(0,keepdim= True)
  # phase 2
  preact = bngain * (preact - meani)/stdi + bnbias
  layer1 = torch.tanh(preact) # layer 1
  layer2 = layer1 @ W2 + b2

  # calculating the running mean and std
  with torch.no_grad():
    mean_running = 0.999 * mean_running + 0.001 * meani
    std_running = 0.999 * std_running + 0.001 * stdi

  # loss function calculation
  loss = F.cross_entropy(layer2, Ytr[mini_batch])

  # resetting gradient to 0
  for p in parameters:
    p.grad = None  # recommended over .zero_() for performance

  # backward pass
  loss.backward()

  # updating this to see our learning rate iteration and loss iteration
  lri.append(i)
  lossi.append(loss.item())

  # update the neural network
  with torch.no_grad(): # Added no_grad() context manager
    for param in parameters :
      lr = 0.1 if i < 100000 else 0.01
      param.data += -lr * param.grad

loss.item()

final_time = time.time() - start_time

print(f'The model took {final_time} seconds to run and the final loss is {loss.item()}')
```

```python
# viewing the loss function now
import matplotlib.pyplot as plt
plt.plot(lri, lossi)
```

```python
# this function replicates the batch norm step
def forward_one_step(context):
    # context: list of tokens length block_size
    emb = C[torch.tensor([context])]  # (1, block_size, d)
    x = emb.view(1, -1)               # (1, in_dim)
    preact = x @ W1 + b1              # (1, h_dim)
    # use running stats for normalization (eval mode)
    rm = mean_running.view(1, -1)
    rstd = std_running.view(1, -1)
    normalized = (preact - rm) / rstd
    post_bn = bngain * normalized + bnbias
    h = torch.tanh(post_bn)
    logits = h @ W2 + b2
    probs = F.softmax(logits, dim=1)
    return probs  # (1, out_dim)
```

```python
# the earlier sampling technique wasn't correct because we're not inferencing the way we were doing it at test
# hence it resulted in gibberish outputs
g = torch.Generator().manual_seed(13)

for _ in range(20):

    out = []
    context = [0] * block_size # block size = 5
    while True:
      probs = forward_one_step(context)
      # emb = C[torch.tensor([context])] # (1,block_size,d)
      # h = torch.tanh(emb.view(1, -1) @ W1 + b1)
      # logits = h @ W2 + b2
      # probs = F.softmax(logits, dim=1)
      ix = torch.multinomial(probs, num_samples=1, generator=g).item()
      context = context[1:] + [ix]
      out.append(ix)
      if ix == 0:
        break

    print(''.join(itow[i] for i in out))
```

### Some questions that I had while doing 'The Fix'
1. **Why do we need so many variables for batch normalisation?**  
Ok let's disect them one by one according to the batch normalisation paper suggestions we need to normalize each batch using the batch mean and std and then eventually multiply with the batch normalisation gain and bias

Screenshot 2025-09-23 at 7.49.44 PM.png


Firstly we have the mean and std for our current batch i.e. meani and stdi  

We also declare the batch normalization gain and bias as 1 and 0 respectively which are to be used for the purpose of scale and shift (see last line in the above image)

But the tricky bit lies wherein we have to calculate the running mean and running std. Why do we need them?

2. **Why do we need the running mean and running std?**   
While doing the inference we cannot calculate the batch mean and std. They would be 0. So to get an idea of the mean and std based on the whole train dataset we have the running mean and std which are calculated using the formula  

***running_mean = 0.99 * running_mean + 0.01 * meani***  
***running_std = 0.99 * running_std + 0.01 * stdi***

The value 0.01 is called momentum and can be adjusted based on the sampling example
